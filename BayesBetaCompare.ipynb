```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def simulateP(p1, n1, p2, n2, nExperiments=100000):
    x1=stats.binom.rvs(p=p1, n=n1, size=nExperiments)
    a1=1+x1
    b1=1+n1-x1

    x2=stats.binom.rvs(p=p2, n=n2, size=nExperiments)
    a2=1+x2
    b2=1+n2-x2

    n=500
    betaRVS_1=[stats.beta.rvs(i, j, size=n) for i,j in zip(a1,b1)]
    betaRVS_2=[stats.beta.rvs(i, j, size=n) for i,j in zip(a2,b2)]

    AvgProb=[i<=j for i,j in zip(betaRVS_1,betaRVS_2)]
    AvgProb=[np.average(i) for i in AvgProb]

    fig=px.histogram(dict(AvgProb=AvgProb), x="AvgProb", nbins=50)
    fig.update_layout(title=dict(text="Distribution of the Average P(rate of X <= rate of Y)", x=0.5))
    fig.update_xaxes(title_text="Probability")
    fig.show()

simulateP(4.2/5, 10, 3.0/5, 18)
```
