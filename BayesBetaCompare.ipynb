```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def simulateP(p1, n1, p2, n2, nExperiments=100000, lev=0.95):
    x1=stats.binom.rvs(p=p1, n=n1, size=nExperiments)
    a1=1+x1
    b1=1+n1-x1

    x2=stats.binom.rvs(p=p2, n=n2, size=nExperiments)
    a2=1+x2
    b2=1+n2-x2

    n=500
    betaRVS_1=[stats.beta.rvs(i, j, size=n) for i,j in zip(a1,b1)]
    betaRVS_2=[stats.beta.rvs(i, j, size=n) for i,j in zip(a2,b2)]

    Probability=[i<=j for i,j in zip(betaRVS_1,betaRVS_2)]
    Probability=[np.average(i) for i in Probability] #Monte Carlo estimation of the posterior P(X<Y|data)

    fig=px.histogram(dict(Probability=Probability), x="Probability", nbins=50)
    fig.update_layout(title=dict(text="Distribution of P(rate of X <= rate of Y)", x=0.5))
    quantil=np.quantile(Probability, lev)
    fig.add_vline(x=quantil, annotation_text=f"{lev*100}% quantile: {quantil*100}%", annotation_position="top left", annotation=dict(font_size=20))
    fig.show()

# p1 p2 are the rate of review, n1 n2 are the number of people giving the review.
simulateP(p1=4.2/5, n1=10, p2=3.0/5, n2=18)
```
